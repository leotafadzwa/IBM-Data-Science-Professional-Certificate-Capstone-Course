```python
#importing libraries
import numpy as np 
import pandas as pd
import json
from pandas.io.json import json_normalize
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
import folium
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
table = pd.read_html(url)
df = table[0]
df.head()
```

```python
df = df[df.Borough != 'Not assigned'] #removing Not assigned.
```

```python
df.Neighborhood.replace('Not assigned',df.Borough,inplace=True) #If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough
```

```python
#grouping
df = df.groupby(['Postal code','Borough'], as_index=False).agg(lambda x: ','.join(x))
df
```
